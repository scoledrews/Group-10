## Research Question: Did the ACA (Affordable Care Act) allow those with lower income to afford medicaid during the years 2013-2016?

```python
#Loading the Data Set into python and importing.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
url = "https://raw.githubusercontent.com/scoledrews/Group-10/refs/heads/main/DATASET%20E.csv?token=GHSAT0AAAAAACZ6YVYK67OMQBCIFUJFLCOWZZI6PJA"
df = pd.read_csv(url)
```

```python
#Finding any missing values from the Data Set
df.isnull()
```

```python
#Focusing on the columns in the dataset
df_focus = df.iloc[ 0:20 , 10:12]
#Median Imputation for Any Missing Values
df_focus['Medicaid Enrollment (2013)'] = df['Medicaid Enrollment (2013)'].fillna(df['Medicaid Enrollment (2013)'].median())
print("\nDataset after Median Imputation for 'Medicaid Enrollment (2013)':")
df_focus
```

```python
#Visualizing the Data Set With a Scatter Plot
df_focus.plot.scatter(x='Medicaid Enrollment (2013)', y='Medicaid Enrollment (2016)')
plt.show()
```

```python
#A Better Visualization
df_subset = df_focus.head(20)

plt.figure(figsize=(10, 5))
plt.plot(df_subset.index, df_subset['Medicaid Enrollment (2013)'], marker='o', label='Medicaid Enrollment 2013', color='blue')
plt.plot(df_subset.index, df_subset['Medicaid Enrollment (2016)'], marker='o', label='Medicaid Enrollment 2016', color='red')

plt.title('Medicaid Enrollment Comparison: 2013 vs 2016')
plt.xlabel('State')
plt.ylabel('Enrollment')

plt.xticks(df_subset.index)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
