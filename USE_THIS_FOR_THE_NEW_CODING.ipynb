```python
#Loading the Data Set into python and importing.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pylab import rcParams
# Import the 'random' module
import random
url = "https://raw.githubusercontent.com/scoledrews/Group-10/refs/heads/main/DATASET%20E.csv?token=GHSAT0AAAAAACZ6YVYK67OMQBCIFUJFLCOWZZI6PJA"
df = pd.read_csv(url)
```

```python
# Display rows where 'State Medicaid Expansion (2016)' is False
df_false=df[df['State Medicaid Expansion (2016)'] == False]
df_false
```

```python
# Display rows where 'State Medicaid Expansion (2016)' is True
df_true=df[df['State Medicaid Expansion (2016)'] == True]
df_true
```

```python
# Split the data into two groups: Medicaid expansion (True) and no expansion (False)
df_true = df[df['State Medicaid Expansion (2016)'] == True]
df_false = df[df['State Medicaid Expansion (2016)'] == False]

# Select 10 random states for Medicaid expansion (True)
random_states_true = random.sample(list(df_true['State']), 10)
df_true_sample = df_true[df_true['State'].isin(random_states_true)]

# Select 10 random states for no Medicaid expansion (False)
random_states_false = random.sample(list(df_false['State']), 10)
df_false_sample = df_false[df_false['State'].isin(random_states_false)]

# Create the bar graph for Medicaid expansion (True)
plt.figure(figsize=(12, 6))

x = np.arange(len(df_true_sample))  # X-axis positions
width = 0.4  # Width of the bars

plt.bar(x - width / 2, df_true_sample['Medicaid Enrollment (2013)'], width=width, label='2013', color='blue')
plt.bar(x + width / 2, df_true_sample['Medicaid Enrollment (2016)'], width=width, label='2016', color='orange')

plt.ylabel("Medicaid Enrollment (In Millions)")  # Y-axis label
plt.xlabel("States")  # X-axis label
plt.title("Medicaid Enrollment in States with Medicaid Expansion (2013-2016)")
plt.xticks(x, df_true_sample['State'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# Create the bar graph for no Medicaid expansion (False)
plt.figure(figsize=(12, 6))

x = np.arange(len(df_false_sample))  # X-axis positions

plt.bar(x - width / 2, df_false_sample['Medicaid Enrollment (2013)'], width=width, label='2013', color='blue')
plt.bar(x + width / 2, df_false_sample['Medicaid Enrollment (2016)'], width=width, label='2016', color='orange')

plt.ylabel("Medicaid Enrollment (In Millions)")  # Y-axis label
plt.xlabel("States")  # X-axis label
plt.title("Medicaid Enrollment in States without Medicaid Expansion (2013-2016)")
plt.xticks(x, df_false_sample['State'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Assuming df_true_sample and df_false_sample are already defined as in the previous code

plt.figure(figsize=(12, 6))

# Plot for Medicaid expansion (True)
plt.plot(df_true_sample['State'], df_true_sample['Medicaid Enrollment (2013)'], marker='o', label='Medicaid Expansion (True) - 2013')
plt.plot(df_true_sample['State'], df_true_sample['Medicaid Enrollment (2016)'], marker='o', label='Medicaid Expansion (True) - 2016')

# Plot for no Medicaid expansion (False)
plt.plot(df_false_sample['State'], df_false_sample['Medicaid Enrollment (2013)'], marker='x', linestyle='--', label='No Medicaid Expansion (False) - 2013')
plt.plot(df_false_sample['State'], df_false_sample['Medicaid Enrollment (2016)'], marker='x', linestyle='--', label='No Medicaid Expansion (False) - 2016')


plt.ylabel("Medicaid Enrollment (In Millions)")
plt.xlabel("States")
plt.title("Medicaid Enrollment Change (2013-2016)")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage increase in Medicaid enrollment for each state
df_true['Percentage Increase'] = ((df_true['Medicaid Enrollment (2016)'] - df_true['Medicaid Enrollment (2013)']) / df_true['Medicaid Enrollment (2013)']) * 100
df_false['Percentage Increase'] = ((df_false['Medicaid Enrollment (2016)'] - df_false['Medicaid Enrollment (2013)']) / df_false['Medicaid Enrollment (2013)']) * 100

# Calculate the average percentage increase for each group
avg_increase_true = df_true['Percentage Increase'].mean()
avg_increase_false = df_false['Percentage Increase'].mean()

# Create the bar graph
plt.figure(figsize=(8, 6))
plt.bar(['Medicaid Expansion (True)', 'No Medicaid Expansion (False)'], [avg_increase_true, avg_increase_false], color=['skyblue', 'lightcoral'])
plt.ylabel('Average Percentage Increase in Medicaid Enrollment')
plt.title('Comparison of Average Percentage Increase in Medicaid Enrollment')
plt.ylim(0, max(avg_increase_true, avg_increase_false) * 1.2) # Adjust y-axis limit for better visualization

# Add percentage values on top of the bars
for i, v in enumerate([avg_increase_true, avg_increase_false]):
    plt.text(i, v + 1, f'{v:.2f}%', ha='center', va='bottom')

plt.show()
```
